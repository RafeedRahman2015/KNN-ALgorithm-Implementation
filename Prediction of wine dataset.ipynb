```
#importing liabraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
from sklearn.datasets import load_wine
wine_dataset=load_wine()
```

```
print("keys of wine datasets",wine_dataset.keys())
```

```
print("Description",wine_dataset["DESCR"])
```

```
print("Feature names:\n", wine_dataset.feature_names)
```

```
print("Target names:", wine_dataset.target_names)
```

```
print("Target:\n", wine_dataset.target)
```

```
from sklearn.model_selection import train_test_split
```

```
x_train,x_test,y_train,y_test=train_test_split(wine_dataset['data'],wine_dataset['target'],test_size=0.25,random_state=0)
```

```
print("x_train shape \n ",x_train.shape)
print("y_train shape \n ",y_train.shape)
```

```
print("x_test shape \n ",x_test.shape)
print("y_test shape \n ",y_test.shape)
```

```
wine_dataframe=pd.DataFrame(x_train,columns=wine_dataset.feature_names)
##creating scatter graphs from datasets
pd.plotting.scatter_matrix(wine_dataframe,c=y_train,figsize=(15,15),marker="x", hist_kwds={'bins': 20}, s=60,alpha=.8)

#wine_dataframe = pd.DataFrame(x_train, columns=wine_dataset.feature_names)

#pd.plotting.scatter_matrix(wine_dataframe, c=y_train, figsize=(15, 15),
 #                          marker='o', hist_kwds={'bins': 20}, s=60,
  #                         alpha=.8)
```

```
fig,ax=plt.subplots()
ax.hist(y_train)
ax.set_xlabel("Class")
ax.set_ylabel("Wine_Samples")
```

```
from sklearn.neighbors import  KNeighborsClassifier
```

```
knn=KNeighborsClassifier(n_neighbors=1)
```

```
knn.fit(x_train,y_train)
```

```
prediction=knn.predict(x_test)
```

```
print(prediction)
print("Predicted Target Names:", wine_dataset['target_names'][prediction])
```

```
fig,ax=plt.subplots()
ax.plot(y_test)
ax.plot(prediction)
ax.set_xlabel("X TEST")
ax.set_ylabel("y TEST AND PREDICTION")

count = 0
for i in range( len(y_test) ):
    if prediction[i] != y_test[i]: 
        count = count + 1
error = count/len(prediction)
print( "Error = %f " % (error*100) + '%' )
accuracy = (1-error)
print( "Accuracy = %f " % (accuracy*100) + '%' )
```
